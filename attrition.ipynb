## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

# additional libraries
from category_encoders import OneHotEncoder
import tensorflow as tf

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
#Viewing the options for X data:
print(attrition_df.columns.unique())
```

```python
# Create a list of at least 10 column names to use as X data
# Create X_df using your selected columns
X_df = attrition_df[['Age','YearsSinceLastPromotion','YearsInCurrentRole',
                     'RelationshipSatisfaction','PerformanceRating','OverTime',
                     'JobInvolvement','JobSatisfaction','JobRole',
                     'DistanceFromHome','HourlyRate','MaritalStatus',
                     'Education','WorkLifeBalance']]

# Show the data types for X_df
print(X_df.dtypes)
```

```python
#finding total number of observations to get a sense for what train and test split ought to be

print(attrition_df.shape[0])
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_df,y_df,test_size=0.2,random_state=42)
```

```python
#finding the number of unique values found in columns OverTime JobRole and MartialStatus

print('OverTime: ', attrition_df.OverTime.nunique())
print('JobRole: ', attrition_df.JobRole.nunique())
print('MaritalStatus: ', attrition_df.MaritalStatus.nunique())

#ok, it looks fine to just use dummies
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_train.head()
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
encoder = OneHotEncoder(cols=['Department'])

# Fit the encoder to the training data
encoder.fit(y_train)

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_dept = encoder.transform(y_train)
y_test_encoded_dept = encoder.transform(y_test)

y_train_encoded_dept = y_train_encoded_dept.drop(columns=["Attrition"])
y_test_encoded_dept = y_test_encoded_dept.drop(columns=["Attrition"])

y_train_encoded_dept
```

```python
# Create a OneHotEncoder for the Attrition column
encoder = OneHotEncoder(cols=['Attrition'])

# Fit the encoder to the training data
encoder.fit(y_train)

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_attrition = encoder.transform(y_train)
y_test_encoded_attrition = encoder.transform(y_test)

y_train_encoded_attrition = y_train_encoded_attrition.drop(columns=["Department"])
y_test_encoded_attrition = y_test_encoded_attrition.drop(columns=["Department"])

y_train_encoded_attrition

```

## Create, Compile, and Train the Model

```python

# Find the number of columns in the X training data
num_cols = X_train_scaled.shape[1]

# Create the input layer
inputs = layers.Input(shape=(num_cols,), name='input_layer')

# Create at least two shared layers
shared_layer_1 = layers.Dense(256, activation='relu', name='shared_layer_1')(inputs)
shared_layer_2 = layers.Dense(128, activation='relu', name='shared_layer_2')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
dept_hidden_layer = layers.Dense(64, activation='relu', name='dept_hidden_layer')(shared_layer_2)

# Create the output layer
dept_output = layers.Dense(3, activation='softmax', name='dept_output')(dept_hidden_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden_layer = layers.Dense(64, activation='relu', name='attrition_hidden_layer')(shared_layer_2)

# Create the output layer
attrition_output = layers.Dense(2, activation='softmax', name='attrition_output')(attrition_hidden_layer)
```

```python
# Create the model
model = Model(inputs=inputs, outputs=[dept_output, attrition_output], name='attrition_model')

# Compile the model
model.compile(optimizer='adam', loss={'dept_output': 'categorical_crossentropy',
                                    'attrition_output': 'binary_crossentropy'},
              metrics={'dept_output': 'accuracy',
                       'attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Need to set run_functions_eagerly to true in order for my model to run. I'm not sure where exactly the issue was, but the fit seems good.

tf.config.run_functions_eagerly(True)
# Train the model
model.fit(X_train_scaled,
          {'dept_output': y_train_encoded_dept,
           'attrition_output': y_train_encoded_attrition},
          epochs=100,
          validation_data=(X_test_scaled,
                           {'dept_output': y_test_encoded_dept,
                            'attrition_output': y_test_encoded_attrition}))
```

```python
# prompt: # Evaluate the model with the testing data

# Evaluate the model with the testing data
model.evaluate(X_test_scaled,
              {'dept_output': y_test_encoded_dept,
               'attrition_output': y_test_encoded_attrition})
```

```python
# prompt: could you generate a confusion matrix of sort on the output of this data?

# Generate predictions for the testing data
y_pred = model.predict(X_test_scaled)

# Extract the predicted classes for the Department and Attrition outputs
dept_pred = np.argmax(y_pred[0], axis=1)
attrition_pred = np.argmax(y_pred[1], axis=1)

# Extract the true classes for the Department and Attrition outputs
dept_true = np.argmax(y_test_encoded_dept.values, axis=1)
attrition_true = np.argmax(y_test_encoded_attrition.values, axis=1)

# Create confusion matrices for the Department and Attrition outputs
from sklearn.metrics import confusion_matrix

dept_cm = confusion_matrix(dept_true, dept_pred)
attrition_cm = confusion_matrix(attrition_true, attrition_pred)

# Print the confusion matrices
print("Department Confusion Matrix:")
print(dept_cm)
print("\nAttrition Confusion Matrix:")
print(attrition_cm)
```
